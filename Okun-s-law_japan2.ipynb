```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# 1. データの取得
data = pd.read_csv('data.csv')

# 2. データの整理
data1 = pd.read_csv('japan-gdp-gross-domestic-product.csv')
gdp_growth = data1["Annual_Change"]

data2 = pd.read_csv('japan-unemployment-rate.csv')
unemployment_rate = data2["Annual_Change"]

# 3. データの解析
# 経済成長と失業率の間の相関
slope, intercept, r_value, p_value, std_err = stats.linregress(gdp_growth, unemployment_rate)

# 4. データの視覚化
# 散布図と回帰直線
plt.axhline(0, color='black', linewidth=0.5) # 横線
plt.axvline(0, color='black', linewidth=0.5) # 縦線
plt.scatter(gdp_growth, unemployment_rate)
plt.plot(gdp_growth.values, intercept + slope*gdp_growth.values, 'r')
plt.xlabel('GDP Growth')
plt.ylabel('Unemployment Rate')
plt.title('Okun\'s Law: Japan')
plt.show()
```
